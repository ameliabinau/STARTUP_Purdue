## Scatter Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

# Create figure object
plt.figure(dpi=200, figsize=(8, 6)) # basic settings of the figure can be edited with plt.figure()

plt.scatter(x, y)
plt.title('Scatter Plot Example') # we can set the title of the plot using plt.title()
plt.xlabel('x-axis') # we can set the labels of the axes using plt.xlabel() and plt.ylabel()
plt.ylabel('y-axis')
plt.show() # plt.show() tells matplotlib that you are finished with the plot and it should be displayed
```

### Customizing Scatter Plots

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8), dpi=150) #plt.subplots allow us to create multiple plots in the same figure. This will be covered later in the notebook! It's very similar to plt. but with different syntax

# Regular scatter plot
axs[0, 0].scatter(x, y)
axs[0, 0].set_title('Regular Scatter Plot')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')

# Scatter plot with edge colors
axs[0, 1].scatter(x, y, edgecolors='black')
axs[0, 1].set_title('Scatter Plot with Edge Colors')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')

# Scatter plot with colormap
scatter = axs[1, 0].scatter(x, y, c=colors, cmap='viridis')
axs[1, 0].set_title('Scatter Plot with Colormap')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')
fig.colorbar(scatter, ax=axs[1, 0])

# Scatter plot with larger markers
axs[1, 1].scatter(x, y, s=200)
axs[1, 1].set_title('Scatter Plot with Larger Markers')
axs[1, 1].set_xlabel('X-axis')
axs[1, 1].set_ylabel('Y-axis')

plt.tight_layout()
plt.show()
```

## Scatter plot with error bars

### Considering the importance of uncertainties in science, knowing how to make a scatter plot with error bars is exceptionally important!

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 10)
y = np.sin(x)
error = np.random.rand(10) * 0.3

plt.figure(dpi=150)
# Plot with error bars
plt.errorbar(x, y, yerr=error, fmt='o', capsize=5, capthick=2)
plt.title('Error Bars Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## Line Plot

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Create figure object
plt.figure(dpi=150, figsize=(8, 6))

plt.plot(x, y)
plt.title('Line Plot Example')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

### The line plot can be customized by changing the line style, color, linewidth, and much more.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Linestyles to be demonstrated
linestyles = ['-', '--', '-.', ':']
titles = ['Solid Line', 'Dashed Line', 'Dash-Dot Line', 'Dotted Line']
colors = ['red', 'green', 'blue', 'orange']
# Plot each line with a different linestyle
for ax, linestyle, title, color in zip(axs.flatten(), linestyles, titles, colors):
    ax.plot(x, y, linestyle=linestyle, color=color, lw=3) # The line width can be changed with lw!
    ax.set_title(title)
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')

plt.tight_layout()
plt.show()
```

## Matplotlib Subplots

One of the most useful features is the ability to put multiple plots all on the same figure!

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1, 0]
y3 = [5, 6, 7, 8, 9, 10]
y4 = [10, 9, 8, 7, 6, 5]

# Create figure object
plt.figure(dpi=150, figsize=(8, 6))


plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Plot 4')

plt.tight_layout()
plt.show()
```

### Sometimes, there is another way to make the same exact plot in matplotlib. Often times one allows more control over the plot than the other.

This example using plt.subplots, rather than plt.subplot from the last example. This is my preferred method to creating subplots in matplotlib.

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]
y2 = [25, 16, 9, 4, 1, 0]
y3 = [5, 6, 7, 8, 9, 10]
y4 = [10, 9, 8, 7, 6, 5]

fig, ax = plt.subplots(2, 2, dpi=150)

ax[0, 0].plot(x, y1)
ax[0, 0].set_title('Plot 1')

ax[0, 1].plot(x, y2)
ax[0, 1].set_title('Plot 2')

ax[1, 0].plot(x, y3)
ax[1, 0].set_title('Plot 3')

ax[1, 1].plot(x, y4)
ax[1, 1].set_title('Plot 4')

plt.tight_layout()
plt.show()
```

## Surface Plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure(dpi=150)
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('3D Surface Plot Example')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

# Advanced Matplotlib

Above you saw the most basic examples and the core functionalities of matplotlib. But, the default settings of matplotlib are not sufficient to make professional grade plots. You can customize the appearance of your plots by changing the default settings of matplotlib. You can change the default settings of matplotlib by using the rcParams dictionary. The rcParams dictionary contains the default settings of matplotlib.

Here's how you can do it:

### 1. Customizing with rcParams

rcParams allow you to set global settings to your matplotlib instance. This means that after running this cell, these settings will be applied to every plot you make until you either change/reset the settings or restart the kernel!

### <span style="color:red">**For the purpose of consistency and clarity of when it is used, I have repeated this code in every cell that I want it to be used in. In practice, you only need to run it once**

```python
import matplotlib.pyplot as plt
import numpy as np

# Set custom parameters
plt.rcParams['font.size'] = 14
plt.rcParams['axes.titlesize'] = 18
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 12
plt.rcParams['lines.linewidth'] = 2

# Create figure object
plt.figure(dpi=150, figsize=(8, 6))

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='Sine Wave')
plt.title('Customized Sine Wave Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

### You can reset matplotlib to its default settings by using `plt.rcdefaults()`

```python
import matplotlib.pyplot as plt

# Set custom parameters
plt.rcParams['font.size'] = 14
plt.rcParams['axes.titlesize'] = 18
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 12
plt.rcParams['lines.linewidth'] = 2

# Create figure object
plt.figure(dpi=150, figsize=(8, 6))

# Create a plot with customized settings
plt.plot([0, 1, 2, 3], [0, 1, 4, 9])
plt.title('Plot with Customized rcParams')
plt.show()

# Reset rcParams to default
plt.rcdefaults()

# Create figure object
plt.figure(dpi=150, figsize=(8, 6))

# Create another plot to show the default settings are restored
plt.plot([0, 1, 2, 3], [0, 1, 4, 9])
plt.title('Plot with Default rcParams')
plt.show()
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'serif' # change default font
rcParams['mathtext.default'] = 'regular'
plt.rcParams.update({'font.size':14}) # change default font size

# change major and minor tick size
rcParams['xtick.major.size'] = 8
rcParams['xtick.minor.size'] = 4
rcParams['ytick.major.size'] = 8
rcParams['ytick.minor.size'] = 4

# Create figure object
plt.figure(dpi=150, figsize=(8, 6))

# Create a plot with customized settings
plt.plot([0, 1, 2, 3], [0, 1, 4, 9])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.minorticks_on()
plt.tick_params(which='both', direction='in', top=True, right=True)

plt.show()
```

### 3. Adding Annotations

Adding annotations with `plt.annotate` can highlight specific points in your plot, making it more informative.

You can also add text in a similar fastion but without the arrow using `plt.text`

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import rcParams
rcParams['font.family'] = 'serif' # change default font
rcParams['mathtext.default'] = 'regular'
plt.rcParams.update({'font.size':14}) # change default font size

# change major and minor tick size
rcParams['xtick.major.size'] = 8
rcParams['xtick.minor.size'] = 4
rcParams['ytick.major.size'] = 8
rcParams['ytick.minor.size'] = 4


x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(dpi=150, figsize=(8, 6))
plt.plot(x, y, label='Sine Wave')
plt.title('Sine Wave with Annotations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add annotation
plt.annotate('Local Max', xy=(1.9, 1), xytext=(3, .9),
             arrowprops=dict(facecolor='black'))

plt.minorticks_on()
plt.tick_params(which='both', direction='in', top=True, right=True)


plt.legend()
plt.show()
```

### 3. Inset plots

While not used incredibly often, inset plots are an extremely useful way to show a zoomed-in version of a specific region of a plot. This can be useful when you want to show a specific region of a plot in more detail. In this example, we will create a simple plot and then add an inset plot to show a zoomed-in version of a specific region of the plot.

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import rcParams
rcParams['font.family'] = 'serif' # change default font
rcParams['mathtext.default'] = 'regular'
plt.rcParams.update({'font.size':14}) # change default font size

# change major and minor tick size
rcParams['xtick.major.size'] = 8
rcParams['xtick.minor.size'] = 4
rcParams['ytick.major.size'] = 8
rcParams['ytick.minor.size'] = 4


# Generate complex data
np.random.seed(0)
x = np.linspace(0, 10, 1000)
y = np.sin(5 * x) + np.random.normal(0, 0.2, len(x))

# Create the main plot
fig, ax = plt.subplots(figsize=(10, 6), dpi=150)
ax.plot(x, y, label='Noisy Sine Wave')
ax.set_title('Main Plot with Inset Zoom')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_xlim(-0.5, 10)

ax.minorticks_on()
ax.tick_params(which='both', direction='in', top=True, right=True)


ax.legend()

# Define the region for the inset (zooming in on x between 2 and 3)
x_inset_start = 2
x_inset_end = 3

# Create inset plot
ax_inset = fig.add_axes([0.55, 0.55, 0.3, 0.3])
mask = (x >= x_inset_start) & (x <= x_inset_end)
ax_inset.plot(x[mask], y[mask], 'r')
ax_inset.set_xlim(x_inset_start, x_inset_end)
ax_inset.set_ylim(min(y[mask]), max(y[mask]))

ax_inset.minorticks_on()
ax_inset.tick_params(which='both', direction='in', top=True, right=True)




plt.show()
```

## Axis Scale

One of the most powerful way of representing difficult data is through changing the axis scale. This is particularly useful when your data spans multiple orders of magnitudes. Using logarithmic axes often make trends in the data much more apparent than they would have been in linear scales.

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import rcParams
rcParams['font.family'] = 'serif' # change default font
rcParams['mathtext.default'] = 'regular'
plt.rcParams.update({'font.size':14}) # change default font size

# change major and minor tick size
rcParams['xtick.major.size'] = 8
rcParams['xtick.minor.size'] = 4
rcParams['ytick.major.size'] = 8
rcParams['ytick.minor.size'] = 4



# Generate complex data
x = np.linspace(1, 100, 100)
y = np.exp(x / 20)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# First subplot with linear scale
ax1.plot(x, y, label='Exponential Growth')
ax1.set_title('Linear Scale')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.legend()
ax1.minorticks_on()
ax1.tick_params(which='both', direction='in', top=True, right=True)

# Second subplot with logarithmic scale
ax2.plot(x, y, label='Exponential Growth')
ax2.set_title('Logarithmic Scale')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis (log scale)')
ax2.set_yscale('log')
ax2.legend()
ax2.minorticks_on()
ax2.tick_params(which='both', direction='in', top=True, right=True)


plt.tight_layout()
plt.show()
```

## Colors and Sizes

It's often the case that you may want to represent 3D or 4D data in a 2D plot. One way to do this is by using colors and/or sizes to represent additional dimensions. In this example, we'll create a scatter plot where the color and size of the points represent two additional dimensions.

Beware: Changing the size of points in a plot can easily lead to a plot that is misleading. Use carefully. Changing the color is almost always prefered. 

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
np.random.seed(0)
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)  # Color mapping
sizes = 1000 * np.random.rand(50)  # Size mapping

# Create scatter plot
plt.figure(figsize=(10, 6), dpi=150)
scatter = plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.colorbar(scatter, label='Color Scale')
plt.title('Scatter Plot with Color and Size Mapping')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show plot
plt.show()
```
